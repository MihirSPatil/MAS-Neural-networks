## Models of a neuron

1.1 An example of the logistic function is defined by 

$\varphi(\nu) = \frac{1}{1 + e^{(-a \nu)}}$

whose limiting values are 0 and 1. Show that the deriviative of $\varphi(\nu)$ with
respect to $\nu$ is given by 

$\frac{d \varphi}{d \nu} = a \varphi(\nu)[1 - \varphi(\nu)]$

What is the value of this derivative at the origin?


```python
sp.init_printing(use_latex=True)
v = sp.Symbol('v')
a = sp.Symbol('a')

phi = 1/(1+sp.exp(-a*v))

derivative_wrt_v = sp.diff(phi,v)

print "Derivative of phi(v):"
print "dphi/dv = {}".format(derivative_wrt_v) 
print "Value of the derivative at the origin:"
print "dphi/dv(0) = {}".format(derivative_wrt_v.subs(v,0))
```
