3.Do the problem 1.13 (Network architecture) from the previous week’s assignment. This time use Python’s (sympy) symbolic toolbox. Finally assume the network presented in fig P1.13 is a binary-classifier, please depict how the input space (R2) is classified on a 2D graph using different colors.

```python
import sympy as sp
import numpy as np
sp.init_printing()
```

```python
def logistic_activation(v):
    return 1/(1+sp.exp(-v))
```

```python
a,b = sp.symbols('a,b')

weight_matrix_1 = np.array([[5,1],[2,-3]])
localised_output_1 = weight_matrix_1.dot([a,b])
output_1 = [logistic_activation(localised_output_1[0]), logistic_activation(localised_output_1[1])]

weight_matrix_2 = np.array([[3,-1],[4,6]])
localised_output_2 = weight_matrix_2.dot(output_1)
output_2 = [logistic_activation(localised_output_2[0]), logistic_activation(localised_output_2[1])]

final_weight_matrix = np.array([1,1])
final_localized_output = final_weight_matrix.dot(output_2)
final_output = logistic_activation(final_localized_output)
final_output
```

```python
%matplotlib inline
f= sp.simplify(final_output)
sp.plotting.plot3d(f)
```

4.
Adjust the data at the "New Classification Example (now *with* bias)" slide, such that a bais becomes necessary (not 0). Validate the perceptron learning algorithm. 

```python
#initial weight with bias as the first parameter of the array
w = np.array([1,0,0])
```

```python
#original data obtained from the slides
x = np.array([1,0,-1])
```

```python
#function to check if product of the weight and data and update the weights accordingly
def check_update(weight, data):
    
    mul = weight.dot(data)
    
    #update the weights only if the product is less than or equal to 0 
    if mul <= 0:
        weight = weight + data
        print("Weight updated")
    else:
        print ("No update to weight performed")
            
        
    return weight
```

```python
print ('initial weight',w)

print ('data:',x)

w = check_update(w,x)

print("final weight",w)
```

```python
#creating new data
d = 1

#randomly generate inputs in the range -1,1 and insert the dummy value to deal with bias
x_new = [np.insert(np.random.randint(-1,2,2),0,d) for _ in range(10)]
```

```python
for i in range(10):
    w = check_update(w,x_new[i])
    print(w)
```

The perceptron learning algorithm is validated as the algorithm converges after a finite number of steps as seen in the above example. Wherein there are no further updates to the weight vector after the sixth iteration, from which we conclude that the algorithm has converged at that time step.

```python
# weights of the network
w0, w1, w2 = sp.symbols('w0,w1,w2')
# input
x1, x2 = sp.symbols('x1,x2')
# desired value
d = sp.Symbol('d')
# learning rate
n = sp.Symbol('n')

y = w0 + x1*w1 + x2*w2
mean_squared_error =  (d-y)**2


diff_w0 = sp.diff(mean_squared_error, w0)
diff_w1 = sp.diff(mean_squared_error, w1)
diff_w2 = sp.diff(mean_squared_error, w2)


# Hence the weights will be updated as follows


print "The weights should be updated based on the following rule"

# New weight =  old weight + learning rate * gradient 
w0 = w0 - n*diff_w0
print "new w0 =", w0
w1 = w1 - n*diff_w1
print "new w0 =", w1
w2 = w2 - n*diff_w2
print "new w0 =", w2
```

```python
# input
x1, x2 = sp.symbols('x1,x2')
# weights of the network
v1, v2 = sp.symbols('v1,v2')
w1, w2 = sp.symbols('w1,w2')
# desired output
d = sp.Symbol('d')
# learning rate
n = sp.Symbol('n')

u1 = w1*x1 + v1*x2
u2 = w2*x1 + v2*x2


y = u1 + u2

mean_squared_error =  (d-y)**2

diff_w1 = sp.diff(mean_squared_error, w1)
diff_w2 = sp.diff(mean_squared_error, w2)
diff_v1 = sp.diff(mean_squared_error, v1)
diff_v2 = sp.diff(mean_squared_error, v2)

# Hence the weights will be updated as follows
# New weight =  old weight + learning rate * gradient 
print ("The weights should be updated based on the following rule")
w1 = w1 - n*diff_w1
print ("new w1 = ", w1)
w2 = w2 - n*diff_w2
print ("new w2 = ", w2)
v1 = v1 - n*diff_v1
print ("new v1 = ", v1)
v2 = v2 - n*diff_v2
print ("new v2 = ", v2)
```
